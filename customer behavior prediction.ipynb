```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('customer_behavior.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Income'] = df['Income'].fillna(df['Income']).mean()
df['MonthlySpending'] = df['MonthlySpending'].fillna(df['MonthlySpending']).mean()
df['SessionDuration'] = df['SessionDuration'].fillna(df['SessionDuration']).mean()
```

```python
df.dropna(subset='SessionDuration', inplace=True)
```

```python
df.isnull().sum()
```

```python
categorical_cols = df.select_dtypes(include='object').columns
categorical_cols
```

```python
df = pd.get_dummies(
    df,
    columns=categorical_cols,
    drop_first=True,
    dtype=int
)
```

```python
numerical_cols = df.select_dtypes(include=['int64','float64']).columns
numerical_cols
```

```python
numerical_features = numerical_cols.drop('PurchaseCategory')
numerical_features
```

```python
scaler = StandardScaler()
df[numerical_features] = scaler.fit_transform(df[numerical_features])
```

```python
X = df.drop('PurchaseCategory', axis=1)
y = df['PurchaseCategory']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
model = DecisionTreeClassifier(
    max_depth=5,
    random_state=42
)
```

```python
scores = cross_val_score(
    model,
    X_train, y_train,
    cv=5,
    scoring='accuracy'
)
```

```python
scores.mean()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```
